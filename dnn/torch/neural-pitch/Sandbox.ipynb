```python
# PTDB Statistics
import numpy as np
import matplotlib.pyplot as pyp
import glob
```

```python
# Dataset Statistics PTDB male/female
dir_f0 = '../../../../ptdb_dataset/SPEECH DATA/combined_reference_f0/'
list_pitch_files = glob.glob(dir_f0 + '*.f0')

list_male_pitches = []
list_female_pitches = []

frac_less = 0

for f in list_pitch_files:
    pitch = np.loadtxt(open(f), delimiter=' ')[:, 0]
    voicing = np.loadtxt(open(f), delimiter=' ')[:, 1]

    if "ref_M" in f:
        list_male_pitches.append(pitch[np.where(voicing!=0)])
    else:
        list_female_pitches.append(pitch[np.where(voicing!=0)])
        # if (np.where(pitch < 125)[0].shape[0] == 0):
            # frac_less = frac_less + 1

list_male_pitches = np.concatenate(list_male_pitches)
list_female_pitches = np.concatenate(list_female_pitches)
```

```python
pyp.subplot(211)
pyp.title('Pitch Histogram of Female Speakers')
pyp.hist(list_female_pitches,bins = 200)
pyp.xlim([0,500])
pyp.subplot(212)
pyp.title('Pitch Histogram of Male Speakers')
pyp.hist(list_male_pitches,bins = 200)
pyp.xlim([0,500])
pyp.tight_layout()
pyp.xlabel('Frequency (Hz)')
```
